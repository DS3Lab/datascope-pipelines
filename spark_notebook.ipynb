## Train model

### Pipe_0

```python
from dspipes import MllibPipelines
from pyspark import SparkContext
from pyspark.sql import SparkSession

sc = SparkContext("local", "First App")
spark = SparkSession(sc)

pipeline = MllibPipelines.create_numerical_pipeline('pipe_1', imputer=False)

start = time.perf_counter()
training = spark.read.load("/Users/david/Projects/rumble_testbed/criteo.kaggle2014.test.parquet")
test = spark.read.load("/Users/david/Projects/rumble_testbed/criteo.kaggle2014.test.parquet")

model = pipeline.fit(training)
prediction = model.transform(test)
predictionAndLabels = prediction.select("label", "prediction") 
correctY = predictionAndLabels.filter("label == prediction").count()
total = predictionAndLabels.count()
print(correctY / total)
```

### Pipe_1

```python
pipeline = MllibPipelines.create_numerical_pipeline('pipe_1', imputer=False)

start = time.perf_counter()
training = spark.read.load("/Users/david/Projects/pipelines/data/uci.parquet")
test = spark.read.load("/Users/david/Projects/pipelines/data/uci_test.parquet")

model = pipeline.fit(training)
prediction = model.transform(test)
result = prediction.select("features", "label", "probability", "prediction") \
    .collect()

for row in result:
    print("features=%s, label=%s -> prob=%s, prediction=%s"
          % (row.features, row.label, row.probability, row.prediction))

end = time.perf_counter()
print(end - start)
```
